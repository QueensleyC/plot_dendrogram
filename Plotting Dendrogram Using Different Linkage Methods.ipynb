```python
import pandas as pd
import seaborn as sns
from scipy.cluster.hierarchy import linkage, dendrogram

import numpy as np
import matplotlib.pyplot as plt
```

# Using a 2-D array of observation vectors

```python
labels = ['A','B','C','D','E','F']
example = np.array([[1,1],
                [1.5,1.5],
                [5,5],
                [3,4],
                [4,4],
                [3,3.5]
                ])
```

## Function for each linkage technique

```python
# Single linkage
def single_linkage(data):
    
    single_clustering = linkage(data, method="single", metric="euclidean", optimal_ordering=True)
    
    # plot dendogram
    fig = plt.figure(figsize=(8, 4))
    plt.title('Single Linkage Hierarchical Clustering Dendrogram')
    plt.xlabel('index')
    plt.ylabel('distance')
    dendrogram(single_clustering,
               labels = labels,  # rotates the x axis labels
               leaf_font_size=8.,  # font size for the x axis labels
              )
    plt.show()
    
    
    
# Average linkage   
def average_linkage(data):
    
    average_clustering = linkage(data, method="average", metric="euclidean", optimal_ordering=True)
    
    # plot dendogram
    fig = plt.figure(figsize=(8, 4))
    plt.title('Average Linkage Hierarchical Clustering Dendrogram')
    plt.xlabel('index')
    plt.ylabel('distance')
    dendrogram(average_clustering,
               labels = labels,  # rotates the x axis labels
               leaf_font_size=8.,  # font size for the x axis labels
              )
    plt.show()
    
    
    
# Complete linkage    
def complete_linkage(data):
    complete_clustering = linkage(data, method="complete", metric="euclidean", optimal_ordering=True)
    
    # plot dendogram
    fig = plt.figure(figsize=(8, 4))
    plt.title('Complete Linkage Hierarchical Clustering Dendrogram')
    plt.xlabel('index')
    plt.ylabel('distance')
    dendrogram(single_clustering,
               labels = labels,  # rotates the x axis labels
               leaf_font_size=8.,  # font size for the x axis labels
              )
    plt.show()
    
```

```python
single_linkage(example)
```

```python
average_linkage(example)
```

```python
complete_linkage(data)
```
